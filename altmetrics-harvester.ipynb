```python
import os
import pandas as pd
os.listdir()
```

```python
df1 = pd.read_json('last_quarter.json')
df1
```

```python
results = pd.read_json( (df1['results']).to_json(), orient='index')
results
```

```python
images = pd.read_json( (results['images']).to_json(), orient='index')
images
```

```python
full_results = pd.concat([results, images], axis=1, join_axes=[results.index])
full_results
```

```python
full_results.to_csv('last_quarter.csv')
os.listdir()
```
