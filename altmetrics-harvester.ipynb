```python
import os
import json
import requests
import pandas as pd
os.listdir()
```

```python
d = json.loads(requests.get('https://api.altmetric.com/v1/citations/1y?doi_prefix=10.5519').text)
del d['query']
with open('one_year.json', 'w') as data_file:
    data = json.dump(d, data_file)
```

```python
df1 = pd.read_json('one_year.json')
df1
```

```python
results = pd.read_json( (df1['results']).to_json(), orient='index')
images = pd.read_json( (results['images']).to_json(), orient='index')
full_results = pd.concat([results, images], axis=1, join_axes=[results.index])
full_results.drop(['cited_by_msm_count', 'cohorts', 'images', 'is_oa', 'tq', 'type'], axis=1, inplace=True)
full_results
```

```python
full_results.to_csv('one_year.csv')
os.listdir()
```
